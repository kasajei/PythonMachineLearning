```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
np.random.seed(12)
for i in range(8):
    x = np.arange(1000)
    y = np.random.randn(1000).cumsum()
    plt.plot(x,y,label=str(i))
plt.legend()
```

```python
import prettyplotlib as ppl
np.random.seed(12)
for i in range(8):
    x = np.arange(1000)
    y = np.random.randn(1000).cumsum()
    ppl.plot(x,y,label=str(i))
ppl.legend()
```

```python
np.random.seed(12)
ppl.pcolormesh(np.random.rand(16,16))
```

```python
import seaborn as sns
```

```python
x1 = np.random.randn(80)
x2 = np.random.randn(80)
x3 = x1 * x2
y1 = .5 + 2 * x1 - x2 + 2.5 * x3 + 3* np.random.randn(80)
y2 = .5 + 2 * x1 - x2 + 2.5 * np.random.randn(80)
y3 = y2 + np.random.randn(80)
```

```python
sns.violinplot([x1,x2,x3])
```

```python
sns.regplot(x2, y2)
```

```python
df = pd.DataFrame(dict(x1=x1, x2=x2, x3=x3, y1=y1, y2=y2, y3=y3))
sns.corrplot(df)
```
